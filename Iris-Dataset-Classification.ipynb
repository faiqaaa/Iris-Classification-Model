```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = pd.read_csv(".\IRIS.csv")
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
sns.FacetGrid(iris, hue="Species", height=6).map(plt.scatter, "Petal.Length", "Sepal.Width").add_legend()
```

```python
flower_mapping = {"setosa": 0, "versicolor": 1, "virginica": 2}
iris["Species"] = iris["Species"].map(flower_mapping)
iris.head()
```

```python
X = iris[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]].values
y = iris["Species"].values
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
predicted = model.predict(X)
predicted
```

```python
# Create a Dataframe with Expected and Predicted values
result = pd.DataFrame({"Expected":y, "Predicted":predicted})
result.iloc[30:56]
```

```python
from sklearn import metrics
print(metrics.classification_report(y, predicted))
```
